# Bayesian Inference

We can use Bayesian Inference to model the distribution of individual repayment statuses and also be able to capture our uncertainty about that distribution so we can better capture the range of values we might see. 

```python
default = df2.pay_1[df2.defaulted == 1]
no_default = df2.pay_1[df2.defaulted  == 0]
n_default = len(default)
n_no_default = len(no_default)
```

```python
default.value_counts()
```

```python
#Initial parameter estimation
alpha_est = np.mean(default)**2 / np.var(default)
beta_est = np.var(default) / np.mean(default)
alpha_est, beta_est
```

```python
#Empirical distribution
_ = plt.hist(no_default, bins=10, alpha=0.5, label='not defaulted clients')
_ = plt.hist(default, bins=10, alpha=0.5, label='defaulted clients')
_ = plt.xlabel('Repayment Status')
_ = plt.ylabel('Frequency')
_ = plt.legend()
```

```python
#Initial simulation
seed(123)
default_model_rvs = gamma(alpha_est, scale=beta_est).rvs(n_default)
```

```python
_ = plt.hist(default_model_rvs, alpha=0.5, label='simulated')
_ = plt.hist(default, alpha=0.5, label='observed')
_ = plt.xlabel('Charge amount')
_ = plt.ylabel('Frequency')
_ = plt.legend()
```

```python
#Create a PYMC3 model
# PyMC3 Gamma seems to use rate = 1/beta
rate_est = 1/beta_est
# Initial parameter estimates we'll use below
alpha_est, rate_est
```

```python
with pm.Model() as model:
    #Create your Gamma distribution with your alpha_ and rate_ stochastic variables and the observed data.
    gamma = pm.Gamma('gamma',alpha=alpha_est,beta=rate_est)
    #Perform 10000 draws.
    trace = pm.sample(10000) #use NUTS for continuous model sampling
```

```python
t = pm.stats.hpd(trace)
```

```python
default = df2.limit_bal[df2.defaulted == 1]
no_default = df2.limit_bal[df2.defaulted  == 0]
n_default = len(default)
n_no_default = len(no_default)
```

```python
#Initial parameter estimation
alpha_est = np.mean(default)**2 / np.var(default)
beta_est = np.var(default) / np.mean(default)
alpha_est, beta_est
```

```python
#Create a PYMC3 model
# PyMC3 Gamma seems to use rate = 1/beta
rate_est = 1/beta_est
# Initial parameter estimates we'll use below
alpha_est, rate_est
```

```python
#Initial simulation
seed(123)
default_model_rvs = gamma(alpha_est, scale=beta_est).rvs(n_default)
```

```python
_ = plt.hist(default_model_rvs, alpha=0.5, label='simulated')
_ = plt.hist(default, alpha=0.5, label='observed')
_ = plt.xlabel('Limit balance')
_ = plt.ylabel('Frequency')
_ = plt.legend()
```

The initial simulation still indicates a much lower limit balance (4000 as compared to the actual 3200) than the observed non-defaulters.

```python
rate_est = 1/beta_est

alpha_est, rate_est
```

```python
with pm.Model() as model:
    #Use an exponential prior for alpha. Call this stochastic variable alpha_.
    alpha_ = pm.Exponential('alpha',alpha_est)
    #Similarly, use an exponential prior for the rate ( 1/𝛽 ) parameter in PyMC3's Gamma. Call this stochastic variable rate_ 
    rate_ = pm.Exponential('beta',beta_est)
    #Create your Gamma distribution with your alpha_ and rate_ stochastic variables and the observed data.
    gamma = pm.Gamma('gamma',alpha=alpha_,beta=rate_,observed=default)
    #Perform 10000 draws.
    trace = pm.sample(10000, step=pm.NUTS()) #use NUTS for continuous model sampling
```

# Bootstrap Sampling

We can try simulating using bootstrap sampling for repayment status and limit balance:

```python
np.random.seed(123)
N_rep = 10000
```

```python
def bs_reps(data,func,size=1):
    
    bs_replicates = np.empty(size)
    
    for i in range(size):
        s = np.random.choice(data,size=len(data))
        bs_replicates[i] = func(s)
    
    return bs_replicates
```

Limit Balance:

```python

bs_replicates = bs_reps(default,np.mean,N_rep)
```

```python
bs_replicates[:10]
```

```python
#95% confidence interval
np.percentile(bs_replicates,[2.5,97.5])
```

This fairs similarly to the result we had for the Bayesian inference model. We can calculate the 95% confidence interval for the difference between the standard deviations of default and non-defaulters (non-defaulters - default). The null hypothesis is that there is no significant difference in the limit balance among default and non-default groups and the alternative hypothesis is that there is a significant difference.




```python
limitbal_std_diff = np.std(no_default) - np.std(default)
limitbal_std_diff
```

```python
#Calculate the differences over 10000 replicates. 
np.random.seed(123)
N_rep = 10000

no_default_replicates = bs_reps(no_default, np.std, N_rep)
default_replicates = bs_reps(default, np.std, N_rep)

diffs = no_default_replicates - default_replicates
diffs
```

```python
bs_std_ci = np.percentile(diffs,[2.5,97.5])
```

```python
_ = plt.figure(figsize=(12,8))
_ = plt.hist(diffs,edgecolor='white',bins=15)
_ = plt.axvline(bs_std_ci[0],color='grey',linestyle='dashed',label="5% CI")
_ = plt.axvline(np.mean(diffs),color='yellow',linestyle='dashed',label="Mean")
_ = plt.axvline(bs_std_ci[1],color='purple',linestyle='dashed',label="95% CI")
_ = plt.xlabel("Limit Balance ($) Standard Deviations")
_ = plt.ylabel("Frequency")
_ = plt.legend()
plt.show()
```

Construct a two-sample bootstrap test:

```python
observed_mean = np.mean(np.concatenate([no_default,default]))
```

```python
observed_mean
```

```python
default_mean_shifted = default - np.mean(default) + observed_mean
no_default_mean_shifted = no_default - np.mean(no_default) + observed_mean
```

```python
bs_replicates_default = bs_reps(default_mean_shifted, np.mean, 10000)
bs_replicates_no_default = bs_reps(no_default_mean_shifted, np.mean, 10000)
```

```python
bs_replicates = bs_replicates_no_default - bs_replicates_default

# Compute and print p-value: p
empirical_diff_means = np.mean(no_default) - np.mean(default)
p = np.sum(bs_replicates >= empirical_diff_means) / len(bs_replicates)

print('p-value =', p)
```

We return a p-value of 0.0 for our two-sample bootrapped hypothesis test which meets the criteria for rejecting the null hypothesis in favor of the alternative which is that limit balances (which are observed as lower for defaulters) are significantly different.
